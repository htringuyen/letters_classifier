# __Letters and Digits Classification__

## Project Description

Create a model that classify letters and digits

We will start with the EMNIST dataset that contain hundreds of thousands of hand-writen digit and letter images categorized into 47 classes

There are many versions of EMNIST dataset, but we will use the balanced version for this project

Here are quick summary of the balanced EMNIST dataset:
> Number of classes: 47
> (0-9, a-z, A-Z but letters C, I, J, K, L, M, O, P, S, U, V, W, X, Y and Z are merged into other classes)

> Number of training samples: 112800

> Number of test samples: 18800
